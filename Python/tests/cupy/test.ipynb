```python
import cupy as cp

# 创建两个随机矩阵
matrix_a = cp.random.rand(3, 4)  # 3x4 矩阵
matrix_b = cp.random.rand(4, 5)  # 4x5 矩阵

# 矩阵乘法
result = cp.dot(matrix_a, matrix_b)

# 打印结果
print("Matrix A:")
print(matrix_a)

print("\nMatrix B:")
print(matrix_b)

print("\nResult of Matrix Multiplication:")
print(result)
```
